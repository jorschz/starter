Aqui está uma explicação breve e resumida de cada conceito relacionado a Python:

1. **`+`, `-`, `*`, `/`**  
Operadores aritméticos: adição, subtração, multiplicação e divisão.

2. **`int` e `float`**  
Tipos numéricos: `int` (números inteiros) e `float` (números decimais).

3. **`//`**  
Divisão inteira: retorna o quociente arredondado para baixo (ex: `7 // 2 = 3`).

4. **`**`**  
Exponenciação: eleva um número a uma potência (ex: `2 ** 3 = 8`).

5. **Operadores bitwise**  
Operam em bits: `&` (AND), `|` (OR), `^` (XOR), `~` (NOT), `<<` (shift left), `>>` (shift right).

6. **Ordem das expressões**  
Prioridade de operadores (ex: multiplicação antes de adição).

7. **Toda expressão tem um valor**  
Qualquer operação/comparação produz um resultado (ex: `3 + 4` vale `7`).

8. **`round()`**  
Arredonda um número (ex: `round(3.7)` → `4`).

9. **`abs()`**  
Retorna o valor absoluto (ex: `abs(-5)` → `5`).

10. **Chamada de função**  
Executa uma função usando parênteses: `nome_função()`.

11. **`%expect` (IPython)**  
Comando mágico para definir saídas esperadas em notebooks.

12. **Erro de sintaxe**  
Código inválido (ex: esquecer `:` ou `)`).

13. **Operações com strings**  
`+` concatena, `*` repete (ex: `"a" * 3` → `"aaa"`).

14. **`len()`**  
Retorna o tamanho de sequências (strings, listas, etc.).

15. **`type()`**  
Retorna o tipo de um dado (ex: `type(10)` → `<class 'int'>`).

16. **Conversões de tipo**  
`int()`, `float()`, `str()` convertem valores.

17. **`TypeError`**  
Erro ao usar tipo incorreto (ex: `"texto" + 5`).

18. **Operandos**  
Valores em uma expressão (ex: em `a + b`, `a` e `b` são operandos).

19. **Linguagens formal vs. natural**  
Código (regras rígidas) vs. linguagem humana (flexível).

20. **Bugs e debugging**  
Bugs = erros; debugging = processo de corrigi-los.

21. **Variáveis e atribuição**  
`nome = valor` cria/atualiza variáveis.

22. **Diagramas de estado**  
Representam valores de variáveis em cada etapa do código.

23. **`import`**  
Carrega módulos (ex: `import math`).

24. **`math`**  
Módulo com funções matemáticas avançadas (ex: `math.sqrt()`).

25. **Statements (comandos)**  
Instruções que executam ações (ex: `if`, `for`, atribuições).

26. **`print()`**  
Exibe valores no console.

27. **Funções built-in**  
Funções nativas do Python (ex: `print()`, `len()`).

28. **Argumentos**  
Valores passados para uma função (dentro dos `()`).

29. **Comentários (`#`)**  
Texto ignorado pelo Python, usado para documentação.

30. **Loop `for`**  
Repete ações sobre itens de uma sequência:  
```python
for item in sequência:
    # código
```

31. **Definição de função (`def`)**  
Cria funções:  
```python
def nome(parâmetros):
    # corpo
```

32. **Parâmetros vs. argumentos**  
Parâmetros: variáveis na definição. Argumentos: valores passados na chamada.

33. **Objeto função**  
Funções são objetos (referenciadas sem `()`).

34. **Chamar funções**  
Executar função com `()` e argumentos.

35. **Contagem começa em 0**  
Índices em Python iniciam em 0 (ex: primeiro elemento é `lista[0]`).

36. **Variáveis e parâmetros são locais**  
Existem apenas dentro da função onde foram criados.

37. **Diagrama de pilha (stack)**  
Mostra frames (contextos) de funções ativas e suas variáveis.

38. **Variáveis globais (`__main__`)**  
Variáveis fora de funções pertencem ao escopo global (`__main__`).

39. **Tracebacks**  
Relatório de erros que mostra a pilha de chamadas.

40. **Citação sobre programação**  
"Programar é o processo de depurar gradualmente até o programa fazer o que você quer."

41. **`for _ in range`**  
`_` é convenção para variável não usada no loop (ex: repetir `n` vezes):  
```python
for _ in range(5):
    print("Olá")
```
